# **Влияние алгоритма ETA на качество доставки**.

## Дизайн эксперимента

### Описание бизнес-проблемы:

- *Текущий прогноз времени доставки для магазинов с быстрой доставкой зафиксирован на уровне 20-60 минут. Разработан алгоритм ETA, который может более точно прогнозировать время доставки в зависимости от разных параметров. Цель состоит в том, чтобы оценить, улучшает ли внедрение алгоритма ETA процесс доставки и бизнес-показатели.*

### Гипотеза:

- *Если мы внедрим алгоритм ETA,  точность прогнозов времени доставки повысится, что приведет к улучшению пользовательского опыта и повышению удовлетворенности клиентов.*

### Какое изменение вносим? Что видит каждая группа?

- *Мы разделим пользователей на две равные группы:*

    - **Контрольная группа** - пользователи по-прежнему будут видеть старый прогноз времени доставки с фиксированным интервалом 20-60 минут.

    - **Тестовая группа** - пользователи получат доступ к новому алгоритму ETA, обеспечивающему более точные прогнозы времени доставки.

### Ожидаемый эффект:

- *Мы ожидаем, что экспериментальная группа, имеющая доступ к новому алгоритму ETA, будет получать более точные прогнозы времени доставки по сравнению с контрольной группой. Это улучшение точности прогнозирования должно привести к улучшению пользовательского опыта и повышению удовлетворенности клиентов.*

### Метрики, по которым будем оценивать эксперимент:

- **Основные метрики:**

    - *Точность времени доставки* - процент заказов с прогнозируемым временем доставки, которые укладываются в фактическое окно доставки.
    
    - *Удовлетворенность клиентов* - измеряется с помощью опросов или рейтингов после покупки.

- **Контр-метрики:**

    - *Коэффициент конверсии* - процент пользователей, которые завершили весь процесс заказа (от выбора продавца до оформления заказа).
    
    - *Объемы заказов* - общее количество заказов, размещенных в течение периода эксперимента.

- **Информационные метрики:**

    - *Вовлеченность пользователей* - количество сеансов на пользователя, средняя продолжительность сеанса.

    - *Время выполнения заказа* - среднее время, затрачиваемое пользователями на завершение заказа.

### При каких развитиях событий и получения каких результатов какие решения я принимаю:

- *Если тестовая группа продемонстрирует значительное улучшение точности времени доставки и удовлетворенности клиентов по сравнению с контрольной группой, мы рассмотрим возможность развертывания нового алгоритма ETA для всех пользователей.*

- *Если в тестовой группе наблюдается снижение коэффициента конверсии или объема заказа, мы выясняем причины этого и вносим необходимые коррективы в алгоритм или пользовательский опыт.*

- *Если в тестовой группе нет значительных улучшений по сравнению с контрольной группой, нам может потребоваться переоценить алгоритм или изучить другие факторы, влияющие на точность времени доставки.*

### Объем выборки для теста:

- *Для оценки размера выборки мы используем калькулятор размера выборки, который можно найти здесь: https://www.evanmiller.org/ab-testing/sample-size.html*

- *Для этого теста мы используем следующие значения:*

    - *Базовый коэффициент конверсии: 50%*

    - *Минимальный обнаруживаемый эффект: 1%*

    - *Статистическая мощность: 80%*

    - *Уровень значимости: 5%*

- *Калькулятор размера выборки показывает, что нам нужно 39,242 пользователей в каждой группе для того, чтобы обнаружить минимальный обнаруживаемый эффект.*

- *Мы увеличим этот объем выборки до 50,000 пользователей в каждой группе, чтобы убедиться, что у нас достаточно данных для обнаружения эффекта.*

### Статистические критерий для каждой метрики:

- *Точность времени доставки* - критерий хи-квадрат или точный критерий Фишера для сравнения доли заказов с точным прогнозом времени доставки между контрольной и экспериментальной группами.

- *Удовлетворенность клиентов* - критерий Стьюдента для независимых выборок или U-критерий Манна-Уитни для сравнения средних показателей удовлетворенности между контрольной и экспериментальной группами.

- *Коэффициент конверсии* - критерий хи-квадрат или точный критерий Фишера для сравнения доли пользователей, завершающих процесс оформления заказа, между контрольной и экспериментальной группами.

- *Объем заказа* - критерий Стьюдента для независимых выборок или U-критерий Манна-Уитни для сравнения среднего объема заказа между контрольной и экспериментальной группами.

- *Вовлеченность пользователей* - t-тест для независимых выборок или U-критерий Манна-Уитни для сравнения средней продолжительности сеанса, рейтинга кликов и других показателей вовлеченности между контрольной и экспериментальной группами.

- *Время выполнения заказа* - t-критерий для независимых выборок или U-критерий Манна-Уитни для сравнения среднего времени выполнения заказа между контрольной и экспериментальной группами.
